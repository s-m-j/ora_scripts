**Задача 1**<br>
Пусть количество страховых событий представляет собой Пуассоновский процесс с годовой интенсивностью 2. Какова вероятность того, что во втором году произойдет хотя бы одно
страховое событие при условии, что в первом произошло 2 страховых события?
<br>
**решение**<br>
Распределение Пуассона описывает поток событий, в котором события наступают независимо друг от друга с некоторой фиксированной средней интенсивностью λ
 (событий в единицу времени). Вероятность случайной величины равной числу событий произошедших за фиксированное время, выражается формулой
\begin{align}
        P(X=k) = \frac{λ^k}{k!}*e^{-λ}
    \end{align}
<br>    
В нашем случае интенсивность λ = 2
<br>
Так как события независимые и число событий во втором году не зависит от предыдущих периодов (периода), нам достаточно вычислить вероятность наступления 0 событий во втором  (и любом другом) году, так как исход 0 событий является "обратным" к интересующему нас исходу "одно и более событий". Вычислим P(X = 0):
\begin{align}
        P(X=0) = \frac{λ^0}{0!}*e^{-2} = \frac{1}{1}*e^{-2} = e^{-2}
    \end{align}

```python
import math
px0 = math.exp(-2)
print(px0)
```

\begin{align}
        P(X>0) = 1 - P(X=0) = 1 - e^{-2}
    \end{align}

```python
import math
px_gt_0 = 1 -math.exp(-2)
print(px_gt_0)
```

**Ответ: ~86%**

**Задача 2**<br>
Пусть 75% клиентов относятся к первой группе надежности, а 25% — ко второй. Пусть каждый год 50% клиентов из первой группы могут перейти во вторую, а 60% клиентов второй группы — в первую. Какова будет доля клиентов в первой группе через 3 года<br>
**Решение**<br>
Обозначим:
- P1(t) - доля клиентов первой группы на шаге t
- P2(t) - доля клиентов второй группы на шаге t
<br>

Из условий задачи:
- P1(0) = 0.75 (начальная доля клиентов в первой группе)
- P2(0) = 0.25 (начальная доля клиентов во второй группе)
- 50% клиентов из первой группы могут перейти во вторую, то есть P_1_2 = 0.5 ·
- 60% клиентов из второй группы могут перейти в первую, то есть P_2_1 = 0.6

<br>
Получим систему уравнений<br>

*   P1(t+1) = (1 - P_1_2) * P1(t) + P_2_1 * P2(t)  
*   P2(t+1) = P_1_2 * P1(t) + (1 - P_2_1) * P2(t)

<br>
решим численно :-)






```python
p1 = 0.75
p2 = 0.25

p12 = 0.5
p21 = 0.6

print('i=%d p1=%.2f p2=%.2f check=%.2f'%(0, p1, p2, p1+p2))
for i in range(8):
  p1_tmp = p1 * (1 - p12) + p2 * p21
  p2_tmp = p2 * (1 - p21) + p1 * p12
  p1 = p1_tmp
  p2 = p2_tmp
  print('i=%d p1=%.5f p2=%.5f check=%.5f'%(i+1, p1, p2, p1+p2))




```

**ответ**<br>
54.25% после 3-го года.
После 5 года процесс станет стационарным

1, а размер убытков от каждого страхового случая распределен равномерно на отрезке [100,150] тыс. руб. Найти математическое
ожидание и среднеквадратическое отклонение общего размера убытков за 3 года.
<br>
**решение**<br>
Для Пуассоновского процесса мат.ожилание = λ = 1 из условий задачи.
<br>
Для равномерного распределения мат ожидание = среднему = 125 из условий задачи
<br>
Из независимости событий вытекает мат ожидание общего размера убытков = 1 * 3 * 125 = 375

Для равномерного распределения<br>
\begin{align}
        D(X) = \frac{(a - b)^2}{12} = 208.333
    \end{align}


```python
D = (100 - 150) ** 2 / 12
print(D)
```
