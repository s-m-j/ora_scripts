–ü—É—Å—Ç—å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–∞—Ö–æ–≤—ã—Ö —Å–æ–±—ã—Ç–∏–π –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è–µ—Ç —Å–æ–±–æ–π –ü—É–∞—Å—Å–æ–Ω–æ–≤—Å–∫–∏–π –ø—Ä–æ—Ü–µ—Å—Å —Å –≥–æ–¥–æ–≤–æ–π –∏–Ω—Ç–µ–Ω—Å–∏–≤–Ω–æ—Å—Ç—å—é 2. –ö–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –≤–æ –≤—Ç–æ—Ä–æ–º –≥–æ–¥—É –ø—Ä–æ–∏–∑–æ–π–¥–µ—Ç —Ö–æ—Ç—è –±—ã –æ–¥–Ω–æ
—Å—Ç—Ä–∞—Ö–æ–≤–æ–µ —Å–æ–±—ã—Ç–∏–µ –ø—Ä–∏ —É—Å–ª–æ–≤–∏–∏, —á—Ç–æ –≤ –ø–µ—Ä–≤–æ–º –ø—Ä–æ–∏–∑–æ—à–ª–æ 2 —Å—Ç—Ä–∞—Ö–æ–≤—ã—Ö —Å–æ–±—ã—Ç–∏—è?

**—Ä–µ—à–µ–Ω–∏–µ**<br>
–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ü—É–∞—Å—Å–æ–Ω–∞ –æ–ø–∏—Å—ã–≤–∞–µ—Ç –ø–æ—Ç–æ–∫ —Å–æ–±—ã—Ç–∏–π, –≤ –∫–æ—Ç–æ—Ä–æ–º —Å–æ–±—ã—Ç–∏—è –Ω–∞—Å—Ç—É–ø–∞—é—Ç –Ω–µ–∑–∞–≤–∏—Å–∏–º–æ –¥—Ä—É–≥ –æ—Ç –¥—Ä—É–≥–∞ —Å –Ω–µ–∫–æ—Ç–æ—Ä–æ–π —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–π —Å—Ä–µ–¥–Ω–µ–π –∏–Ω—Ç–µ–Ω—Å–∏–≤–Ω–æ—Å—Ç—å—é Œª
 (—Å–æ–±—ã—Ç–∏–π –≤ –µ–¥–∏–Ω–∏—Ü—É –≤—Ä–µ–º–µ–Ω–∏). –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã —Ä–∞–≤–Ω–æ–π —á–∏—Å–ª—É —Å–æ–±—ã—Ç–∏–π –ø—Ä–æ–∏–∑–æ—à–µ–¥—à–∏—Ö –∑–∞ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–µ –≤—Ä–µ–º—è, –≤—ã—Ä–∞–∂–∞–µ—Ç—Å—è —Ñ–æ—Ä–º—É–ª–æ–π
\begin{align}
        P(X=k) = \frac{Œª^k}{k!}*e^{-Œª}
    \end{align}
<br>    
–í –Ω–∞—à–µ–º —Å–ª—É—á–∞–µ –∏–Ω—Ç–µ–Ω—Å–∏–≤–Ω–æ—Å—Ç—å Œª = 2
<br>
–¢–∞–∫ –∫–∞–∫ —Å–æ–±—ã—Ç–∏—è –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã–µ –∏ —á–∏—Å–ª–æ —Å–æ–±—ã—Ç–∏–π –≤–æ –≤—Ç–æ—Ä–æ–º –≥–æ–¥—É –Ω–µ –∑–∞–≤–∏—Å–∏—Ç –æ—Ç –ø—Ä–µ–¥—ã–¥—É—â–∏—Ö –ø–µ—Ä–∏–æ–¥–æ–≤ (–ø–µ—Ä–∏–æ–¥–∞), –Ω–∞–º –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –≤—ã—á–∏—Å–ª–∏—Ç—å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –Ω–∞—Å—Ç—É–ø–ª–µ–Ω–∏—è 0 —Å–æ–±—ã—Ç–∏–π –≤–æ –≤—Ç–æ—Ä–æ–º  (–∏ –ª—é–±–æ–º –¥—Ä—É–≥–æ–º) –≥–æ–¥—É, —Ç–∞–∫ –∫–∞–∫ –∏—Å—Ö–æ–¥ 0 —Å–æ–±—ã—Ç–∏–π —è–≤–ª—è–µ—Ç—Å—è "–æ–±—Ä–∞—Ç–Ω—ã–º" –∫ –∏–Ω—Ç–µ—Ä–µ—Å—É—é—â–µ–º—É –Ω–∞—Å –∏—Å—Ö–æ–¥—É "–æ–¥–Ω–æ –∏ –±–æ–ª–µ–µ —Å–æ–±—ã—Ç–∏–π". –í—ã—á–∏—Å–ª–∏–º P(X = 0):
\begin{align}
        P(X=0) = \frac{Œª^0}{0!}*e^{-2} = \frac{1}{1}*e^{-2} = e^{-2}
    \end{align}

```python
import math
px0 = math.exp(-2)
print(px0)
```

\begin{align}
        P(X>0) = 1 - P(X=0) = 1 - e^{-2}
    \end{align}

```python
import math
px_gt_0 = 1 -math.exp(-2)
print(px_gt_0)
```

**–û—Ç–≤–µ—Ç: ~86%**

**–ó–∞–¥–∞—á–∞ 2**<br>
–ü—É—Å—Ç—å 75% –∫–ª–∏–µ–Ω—Ç–æ–≤ –æ—Ç–Ω–æ—Å—è—Ç—Å—è –∫ –ø–µ—Ä–≤–æ–π –≥—Ä—É–ø–ø–µ –Ω–∞–¥–µ–∂–Ω–æ—Å—Ç–∏, –∞ 25% ‚Äî –∫–æ –≤—Ç–æ—Ä–æ–π. –ü—É—Å—Ç—å –∫–∞–∂–¥—ã–π –≥–æ–¥ 50% –∫–ª–∏–µ–Ω—Ç–æ–≤ –∏–∑ –ø–µ—Ä–≤–æ–π –≥—Ä—É–ø–ø—ã –º–æ–≥—É—Ç –ø–µ—Ä–µ–π—Ç–∏ –≤–æ –≤—Ç–æ—Ä—É—é, –∞ 60% –∫–ª–∏–µ–Ω—Ç–æ–≤ –≤—Ç–æ—Ä–æ–π –≥—Ä—É–ø–ø—ã ‚Äî –≤ –ø–µ—Ä–≤—É—é. –ö–∞–∫–æ–≤–∞ –±—É–¥–µ—Ç –¥–æ–ª—è –∫–ª–∏–µ–Ω—Ç–æ–≤ –≤ –ø–µ—Ä–≤–æ–π –≥—Ä—É–ø–ø–µ —á–µ—Ä–µ–∑ 3 –≥–æ–¥–∞<br>
**–†–µ—à–µ–Ω–∏–µ**<br>
–û–±–æ–∑–Ω–∞—á–∏–º:
- P1(t) - –¥–æ–ª—è –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–µ—Ä–≤–æ–π –≥—Ä—É–ø–ø—ã –Ω–∞ —à–∞–≥–µ t
- P2(t) - –¥–æ–ª—è –∫–ª–∏–µ–Ω—Ç–æ–≤ –≤—Ç–æ—Ä–æ–π –≥—Ä—É–ø–ø—ã –Ω–∞ —à–∞–≥–µ t
<br>

–ò–∑ —É—Å–ª–æ–≤–∏–π –∑–∞–¥–∞—á–∏:
- P1(0) = 0.75 (–Ω–∞—á–∞–ª—å–Ω–∞—è –¥–æ–ª—è –∫–ª–∏–µ–Ω—Ç–æ–≤ –≤ –ø–µ—Ä–≤–æ–π –≥—Ä—É–ø–ø–µ)
- P2(0) = 0.25 (–Ω–∞—á–∞–ª—å–Ω–∞—è –¥–æ–ª—è –∫–ª–∏–µ–Ω—Ç–æ–≤ –≤–æ –≤—Ç–æ—Ä–æ–π –≥—Ä—É–ø–ø–µ)
- 50% –∫–ª–∏–µ–Ω—Ç–æ–≤ –∏–∑ –ø–µ—Ä–≤–æ–π –≥—Ä—É–ø–ø—ã –º–æ–≥—É—Ç –ø–µ—Ä–µ–π—Ç–∏ –≤–æ –≤—Ç–æ—Ä—É—é, —Ç–æ –µ—Å—Ç—å P_1_2 = 0.5 ¬∑
- 60% –∫–ª–∏–µ–Ω—Ç–æ–≤ –∏–∑ –≤—Ç–æ—Ä–æ–π –≥—Ä—É–ø–ø—ã –º–æ–≥—É—Ç –ø–µ—Ä–µ–π—Ç–∏ –≤ –ø–µ—Ä–≤—É—é, —Ç–æ –µ—Å—Ç—å P_2_1 = 0.6

<br>
–ü–æ–ª—É—á–∏–º —Å–∏—Å—Ç–µ–º—É —É—Ä–∞–≤–Ω–µ–Ω–∏–π<br>

*   P1(t+1) = (1 - P_1_2) * P1(t) + P_2_1 * P2(t)  
*   P2(t+1) = P_1_2 * P1(t) + (1 - P_2_1) * P2(t)

<br>
—Ä–µ—à–∏–º —á–∏—Å–ª–µ–Ω–Ω–æ :-)






```python
p1 = 0.75
p2 = 0.25

p12 = 0.5
p21 = 0.6

print('i=%d p1=%.2f p2=%.2f check=%.2f'%(0, p1, p2, p1+p2))
for i in range(8):
  p1_tmp = p1 * (1 - p12) + p2 * p21
  p2_tmp = p2 * (1 - p21) + p1 * p12
  p1 = p1_tmp
  p2 = p2_tmp
  print('i=%d p1=%.5f p2=%.5f check=%.5f'%(i+1, p1, p2, p1+p2))




```

–ü—É—Å—Ç—å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–∏–∑–æ—à–µ–¥—à–∏—Ö —Å—Ç—Ä–∞—Ö–æ–≤—ã—Ö —Å–ª—É—á–∞–µ–≤ –º–æ–¥–µ–ª–∏—Ä—É–µ—Ç—Å—è —Å –ø–æ–º–æ—â—å—é —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ü—É–∞—Å—Å–æ–Ω–∞ —Å –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–º 6, –∞ —É—â–µ—Ä–± –æ—Ç –∫–∞–∂–¥–æ–≥–æ —Å–æ–±—ã—Ç–∏—è —Å –ø–æ–º–æ—â—å—é —ç–∫—Å–ø–æ–Ω–µ–Ω—Ü–∏–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å–æ —Å—Ä–µ–¥–Ω–∏–º –∑–Ω–∞—á–µ–Ω–∏–µ–º 30000 —Ä—É–±. –ù–∞–π—Ç–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —É–±—ã—Ç–∫–∞ –Ω–µ –±—É–¥–µ—Ç
